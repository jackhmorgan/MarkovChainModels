```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
title = 'Dynamic Credit Risk Results on Hardware'
filename = title.replace(" ", "")+'.png'
categories = ['$\\mathcal{L} = 0$', '$\\mathcal{L} = 1$', '$\\mathcal{L} = 2$', '$\\mathcal{L} = 3$']
data_set1 = [0.06349, 0.07945, 0.098127, 1.0]
data_set2 = [0.3, 0.09, 0.19, 1.0]
data_set3 = [0.18, 0.21, 0.33, 0.80]

# Set up positions for bars on x-axis
bar_width = 0.25
bar_positions_set1 = np.arange(len(categories))
bar_positions_set2 = bar_positions_set1 + bar_width
bar_positions_set3 = bar_positions_set2 + bar_width

fig = plt.figure()

# Plotting the bar graphs
plt.bar(bar_positions_set1, data_set1, width=bar_width, label='Ideal Probability')
plt.bar(bar_positions_set2, data_set2, width=bar_width, label='Honeywell H1')
plt.bar(bar_positions_set3, data_set3, width=bar_width, label='IBM Torino')

# Adding labels and title
plt.xlabel('Loss Value')
plt.ylabel('$\\mathcal{P} \\ \\left(\mathcal{L} \\leq X \\right)$')
plt.title(title)
plt.xticks(bar_positions_set2, categories)  # Set x-axis tick positions to the middle of each group of bars
plt.legend()

# Show the plot
plt.show()
fig.savefig(filename)
```
